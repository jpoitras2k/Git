Decision Trees (DTs) are a non-parametric supervised learning method used for classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features. A tree can be seen as a piecewise constant approximation.

<br>
A few notes on Decision Trees:
<br>- They are generally simple to use, and understand because they can be visualized. I.e a white box model, not a black box!
<br>- Prone to creating overly-complex models, what we call overfitting.
<br>- Prone to poor stability because a change in sample data can drastically change the tree model. 
<br>(ScikitLearn)

## Read data from csv file and perform preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Social_Network_Ads.csv")
df_visualize = pd.read_csv("Social_Network_Ads.csv").head(20)

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
```

```python
# Create X and y data
x_columns = 2
X = df.iloc[0:20, 0:x_columns].values
y = df.iloc[0:20, x_columns].values
```

## Decision Tree Classifier - ON A SMALL DATASET

A few comments about Decisions Trees:
- Beneficial to use Decision Trees when: 
     - the data has a non-linear trend and extrapolation outside the training data is not important.
     - Can struggle with large datasets. It's fast to train, but can be slow to predict.
   
- Not ideal for:
    - time series forecasting. Time based data requires the ability to identify a increasing/decresing trends, which DT's and RF's don't do very well
    - imbalanced datasets
    - high feature count datasets with low contribution (noise in the forest)

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier(criterion = "entropy", 
                                    max_leaf_nodes = 20)
```

```python
model_tree.fit(X, y)
```

```python
y_pred_tree = model_tree.predict(X)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y, y_pred_tree))
print(classification_report(y,y_pred_tree))
```

## Now run the Decision Tree Pipeline with the Whole Dataset

```python
# Create X and y data
x_columns = 2
X = df.iloc[:, 0:x_columns].values
y = df.iloc[:, x_columns].values

# Create Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)

# Choose the model
model_tree = DecisionTreeClassifier(criterion = "entropy", 
                                    max_leaf_nodes = 10)
# Train the model
model_tree.fit(X_train, y_train)

#Evaluate the model
y_pred_tree = model_tree.predict(X_test)
print(confusion_matrix(y_test,y_pred_tree))
print(classification_report(y_test,y_pred_tree))

#Visualize the model
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (12,12), dpi = 300)
plot_tree(model_tree);
```

```python
# Plot the larger Decision Tree (with help from chatGPT!)
sn.scatterplot(x=X[:, 0], y=X[:, 1], hue=y,  s=60,)
plt.xlabel('Age'), plt.ylabel('Est. Salary')

# Recursive function to extract and plot splits
def plot_splits(tree, node_id=0, x_min=None, x_max=None, y_min=None, y_max=None):
    # Feature index and threshold for current split
    if tree.feature[node_id] == -2:
        return  # leaf node, no split

    f = tree.feature[node_id]
    t = tree.threshold[node_id]

    # Default bounds
    if x_min is None: x_min = X[:, 0].min() - 1
    if x_max is None: x_max = X[:, 0].max() + 1
    if y_min is None: y_min = X[:, 1].min() - 1
    if y_max is None: y_max = X[:, 1].max() + 1

    # Vertical split (Feature 0)
    if f == 0:  # vertical split
        plt.plot([t, t], [y_min, y_max], color='purple', linestyle='--', linewidth=2)
        plot_splits(tree, tree.children_left[node_id], x_min, t, y_min, y_max)
        plot_splits(tree, tree.children_right[node_id], t, x_max, y_min, y_max)
    
    elif f == 1:  # horizontal split (Feature 1)
        plt.plot([x_min, x_max], [t, t], color='red', linestyle='--', linewidth=2)
        plot_splits(tree, tree.children_left[node_id], x_min, x_max, y_min, t)
        plot_splits(tree, tree.children_right[node_id], x_min, x_max, t, y_max)

# Plot
plot_splits(model_tree.tree_)

plt.title('Decision Tree Splits')
plt.show()
```

## Random Forest Classifier

Random Forests build on the concept of the Decision Tree. However, instead of one tree incorporating all the data and all the features, an 'Ensemble' of trees is generated with different groupings of features and sample data.   

```python
from sklearn.ensemble import RandomForestClassifier
model_RF = RandomForestClassifier(n_estimators = 10, # Default is 100, but we have small dataset
                                  bootstrap = True) # If false, then all samples will be used in each tree  
```

```python
model_RF.fit(X_train, y_train)
y_pred_RF = model_RF.predict(X_test)
model_RF.score(X_test, y_test)
```

```python
print(confusion_matrix(y_test,      y_pred_RF))
print(classification_report(y_test, y_pred_RF))
```

The Random Forest outperforms the Decision Tree even with only two features because:
- multiple creations of the tree are being considered inside the algorithm, which is not done in the DT. Can test this by trying n_estimators = 1, which then performs similarily to DT. The results of all the trees then 'vote' on the class prediction.
- Different subsets of the training data are used on the various trees. Unless n_estimators = 1 and bootstrap = False, then all the sample points will be used.
- In datasets with more columns, not every feature will be used in every tree, a subset of features is selected from at each node.
- 
(Note: Our score is sensitive because we have a small, imbalanced dataset).

Let's add a few more arguments to our modelling. There are many more, check out the Random Forest article on Sklearn: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

```python
model_RF = RandomForestClassifier(n_estimators = 10, # Default is 100
                                  criterion = "gini", # Can use {gini, entropy, log_loss}
                                  bootstrap = True, # True = use a subset of samples, False = use all samples
                                  max_features = 'sqrt', # This is the number of randomly chosen features to decide between at each split
                                  class_weight = "balanced") # or a dictionary with key:value pair set-up: {class_label: weight}

model_RF.fit(X_train, y_train)
model_RF.score(X_test, y_test)
```
