```python
def main():

    from sklearn.datasets import load_iris
    X, y = load_iris(return_X_y=True)

    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

    ### Write your code here. Do not change any other parts of the code  
    from sklearn.neighbors import KNeighborsClassifier 
    # Number of Neighbours, Choose value of K 
    K = 3
    model = KNeighborsClassifier(n_neighbors=K)
    model.fit(X_train, y_train)

    #  Predict the test using X_test
    y_pred = model.predict(X_test)
    #print (y_pred)

    # Score the test
    score= model.score(X_test, y_test)
    print(score)

    ### Write code upto here  
    #uncomment the following two lines when you write the code for them to run   
    #print(score)
    #return score
main()
```
