Visualizations are the name collectively given to the various ways that we can display data as images as opposed to numerical or categorical data.

A few examples of visualizations are scatter plots, bar plots and heat maps.

This notebook will walk you through making some basic scatterplot visualizations including identifying trends using the Matplotlib Library.

### Data Points
When we collect data, we can plot it on a chart using it's values. Let's take the data point shown below, In this data point, the values in positions X1 and y1 represent two measurements of some kind for an opject. 

```python
X1 = 2
y1 = 21
```

 We can say that these two measurements for this object make a 'data point':

```python
point_1 = [X1, y1]
point_1
```

If we want to show this data point on a plot, we can create a visualization:

```python
import matplotlib.pyplot as plt

# Top plot using individual values
plt.scatter(X1, y1)
plt.show()

# Lower plot made using the data extracted from the point
plt.scatter(point_1[0], point_1[1])
plt.show()

#These two plots show the exact same thing because the data is the same 
```

The x value corresponds to the x-axis (horizontal axis) and the y value corresponds to the y-axis (vertical axis). The point is plotted at the intersection of the two values on their corresponding axis. In this examples, point(2, 21) plots above the 2 on the horizontal axis and to the right of the 21 on the vertical axis. 

```python
plt.scatter(point_1[0], point_1[1])
plt.xlabel("Horizontal Axis - Independent Variable")
plt.ylabel("Vertical Axis - Dependent Variable")
plt.xlim(0,3), plt.ylim(0,30)

plt.show() # Forces the notebook to show the output
```

### Scatter Plots
The most common way of visualizing the relationships between data, is to use a scatter plot. In a scatter plot, we have an independent variable on the x-axis and a dependent variable on the y-axis. The idea being, we want to see if the independent variable has a direct influence (correlation) to the dependent variable.

If we were to collect measurements for additional objects in the data sets, we will have more data points. Here we've added 4 more measurements to our original data set (stared as the single point). I added all the values for one measurement to the X list and all the values for the second measurement to the y list. <br> *Note: The values must be in the same order!*

```python
# X Measurements:
X1 = 2
X2 = 3
X3 = 4
X4 = 5
X5 = 2.5

# y Measurements:
y1 = 21
y2 = 32
y3 = 39
y4 = 50
y5 = 24

X_values = [X1, X2, X3, X4, X5]
y_values = [y1, y2, y3, y4, y5]

print("X-values = ",X_values,"\ny-values = ",y_values)
```

We can now plot this series of data points in a scatter plot to visualize if there is a relationship to the data. In other words, if I change the independant variable, is there a direct and obvious influence on the dependent variable? Let's see!

```python
# This time instead of using the single point, I can plot all the 
# points at the same time because they are listed in order 
# in X_values and y_values lists

plt.scatter(X_values, y_values)
plt.xlabel("Horizontal Axis - Independent Variable")
plt.ylabel("Vertical Axis - Dependent Variable")
plt.xlim(0,6), plt.ylim(0,60)

plt.show()
```

There is a relationship! When the value of X is increased, so is the value of y. Therefore, we can say that the y value is dependent on the X value. In this case, because it looks like we could draw a line straight through all the points, we could also say it is highly, positively, linearly correlated. 

```python
plt.scatter(X_values, y_values)
plt.plot(range(1,7,1), range(10,70,10), color="red")
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(0,6), plt.ylim(0,60)

plt.show()
```

We can also plot negative numbers on a scatter plot:

```python
X_values = [-9, -2, 0, 4, 9]
y_values = [8, 4, 0, -5, -8]

plt.axhline(color = "black"), plt.axvline(color = "black")
plt.scatter(X_values, y_values, s = 100)
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(-10,10), plt.ylim(-10,10)
```

This would be an example of a negatively, highly correlated variable. In other words, as the X value increases, the y value decreases. We can see this by adding a trend line in red.

```python
plt.axhline(color = "black"), plt.axvline(color = "black")
plt.scatter(X_values, y_values, s = 100)
plt.plot([-9.5,9.5], [9.5,-9.5], color="red")
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(-10,10), plt.ylim(-10,10)
```

Adding the trend line helps us make predictions of where new data could fit in our relationship. <br> Let's say we found a new object and made a measurement for x, but couldn't get the y value without damaging the object in some way or it is really expensive. How can we get the new y? By using the linear trend we see in the data!

In this example, if our new x is 6, we can estimate the new y value at -6 because that is where the trend line intersects. We could also use the points to calculate the slope of the line and the y-intercept and then plug the new X value into the equation, like described in the accompanying excel document.

```python
plt.axhline(color = "black"), plt.axvline(color = "black")
plt.scatter(X_values, y_values, s = 100)
plt.plot([-9.5,9.5], [9.5,-9.5], color="red")
plt.scatter(6, -6, s = 100, color = "green")
plt.plot([6,6], [-6,-10], color="green"), plt.plot([-10,6], [-6,-6], color="green")
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(-10,10), plt.ylim(-10,10)
```

We can do one better and actually calculate the line of best fit instead of eyeballing it!

```python
#find line of best fit
import numpy as np
X_values = np.array([-9, -2, 0, 4, 9])

m, b = np.polyfit(X_values, y_values, 1)
print(m, b)
```

```python
plt.axhline(color = "black"), plt.axvline(color = "black")
plt.scatter(X_values, y_values, s = 100)
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(-10,10), plt.ylim(-10,10)
plt.plot(X_values, m*X_values+b, color='red', linestyle='--', linewidth=2)
```

And calculate the actual estimate of the new object with measurement X = 6:

```python
X_new = 6
y_new = m*X_new+b

print(y_new) # Pretty close to our eyeball estimate!
```

In machine learning, many methods try and find an equation of a line (or higher order equation) to estimate new values like we just did here, but in much more complex datasets. <br><br>For example, this dataset has 3 variables: X, Y & z. 

```python
ax = plt.axes(projection='3d') # Specifically create a 3D axis named 'ax'
ax.set_xlabel("Independent Variable (X)")
ax.set_ylabel("Independent Variable (Y)")
ax.set_zlabel("Dependent Variable (z)")

# Create and plot X, Y, z data
plt.scatter(range(0,10,1), range(1,100,10), range(0,1000, 100)) 
```

Visualizing dimensions beyond 3 becomes tricky, if not impossible, and that is why we typically work with subsets of the dataset to look for relationships.
