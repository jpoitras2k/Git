This notebook uses some common data exploration techniques, including Matplotlib and Seaborn visualizations to explore the GPA Dataset.

## Import Libraries

```python
import pandas as pd
import numpy as np

#import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sn
```

## Read csv file into a dataframe

```python
df = pd.read_csv("gpa.csv")
```

```python
df
```

```python
df.describe()
```

## Calculate Summary Statistics

```python
#Calculate the average of each column in the dataset
df.mean()
```

```python
# Determine the middle value of each column

df.median()
```

```python
# Determine the most commonly occurring value in each column

# high_GPA column has two MODES, 3.28 & 3.42 as both numbers appear 4 times in the column (axis = 0)
# univ_GPA has only one MODE, 3.28, it appears 7 times!

df.mode(axis=0)
```

```python
df["univ_GPA"].value_counts()
```

```python
# Calculate the standard deviation of each attribute (column)

df.std()
```

```python
# Calculate the variance of each attribute

df.var()
```

```python
# Determine the correclation coefficiant between the attributes, the higher 
# the correlation (closer to 1.0), the more related the values are by a mathematical equation
# .corr() calculates r, which is the correlation coefficient between the two variables

df.corr()
```

### Other helpful exploration commands

```python
# Return unique values in a column
df["high_GPA"].unique()
```

```python
# Determine number of unique values by counting the length of the returned string
len(df["high_GPA"].unique())
```

```python
# Determine number of unique values by using the built in size function
df["high_GPA"].unique().size
```

### Add some visualizations

```python
# Basic Scatter plot
plt.scatter(x=df["high_GPA"], y=df["univ_GPA"])
```

```python
# Add plot details
plt.scatter(df["high_GPA"],df["univ_GPA"], color="purple")
plt.title("Predicting University Grades \n from High School GPA", size = 18)
plt.xlabel("High School (GPA)")
plt.ylabel("University (GPA)")
plt.show()
```

```python
#Create a histogram of the distribution of the grades
plt.hist(df)
plt.show()
```

```python
# Add additional chart features
years = ["High School", "University"]

plt.figure(figsize=(12,4))
plt.hist(df, label = years, bins = 20)
plt.title("Distribution of School Grades", size = 15)
plt.xlabel("Grade Point Average")
plt.ylabel("Count")
plt.legend(loc='upper left', frameon=False)
```

```python
# Add a line plot
plt.figure(figsize=(12,4))
plt.plot(df["high_GPA"], df["univ_GPA"])
plt.title("Variation of High School Grades", size = 15)
plt.xlabel("High School Grade Point Average")
plt.ylabel("University Grade Point Average")
plt.show()
```

```python
# Line charts usually show 'ordered data", like timeseries data, so we can mimic this by sorting on the X column:
df = df.sort_values(by = "high_GPA")

plt.figure(figsize=(12,4))
plt.plot(df["high_GPA"], df["univ_GPA"])
plt.title("Variation of High School Grades", size = 15)
plt.xlabel("High School Grade Point Average")
plt.ylabel("University Grade Point Average")
plt.show()
```

```python
# Create box plots with the Seaborn Library
plt.figure(figsize=(12,4))
sn.boxplot(data = df)
```

```python
# Create violin plots with the Seaborn Library - these give additional 
# insight into the behaviour of the dataset.

plt.figure(figsize=(12,4))
sn.violinplot(data = df, palette=("rocket")) 
# modify colour palette: https://seaborn.pydata.org/tutorial/color_palettes.html
```

```python
# Create a Pair Plot of all the attribute columns to quickly visualize the relationships 
# instead of making many individual scatter plots

sn.pairplot(df)
```

```python
# Subplots allow us to combine plots into one grid space
fig, axs = plt.subplots(2, 2, figsize=(11, 8))
fig.suptitle('GPA Comparisons', fontsize=16)

# Fill the grid spaces with some charts:
# A Distribution plot (histogram of University GPA's using Seaborn)
sn.histplot(df['univ_GPA'] , bins=14, kde=True, ax=axs[0, 0], color='purple')
axs[0, 0].set_title('Distribution Plot (univ_GPA)'), axs[0, 0].set_xlabel('University GPA Distribution')

# A Scatter plot:
axs[0, 1].scatter(df['high_GPA'], df['univ_GPA'], color='darkviolet', alpha=0.8)
axs[0, 1].set_title('Scatter Plot'), axs[0, 1].set_xlabel('high_GPA'), axs[0, 1].set_ylabel('univ_GPA')

# A Distribution plot (histogram of High School GPA using Seaborn)
sn.histplot(df['high_GPA'] , bins=14, kde=True, ax=axs[1, 0], color='thistle')
axs[1, 0].set_title('Distribution Plot (high_GPA)'), axs[1, 0].set_xlabel('High School GPA Distribution')

# A Bar plot (average GPA by bins of high_GPA)
# Cut slices the pandas dataframe into n bins by the value range in the stated column 
df['GPA_bin'] = pd.cut(df['high_GPA'], bins=5) 

# The values in the bin are then aggregrated to the mean university GPA in that bin
gpa_means = df.groupby('GPA_bin', observed = False)[['high_GPA', 'univ_GPA']].mean().reset_index()

axs[1, 1].bar(gpa_means['GPA_bin'].astype(str), gpa_means['univ_GPA'], color='mediumorchid')
axs[1, 1].set_title('Bar Plot: Avg univ_GPA by high_GPA Bin')
axs[1, 1].set_xlabel('high_GPA Bin'), axs[1, 1].set_ylabel('Average univ_GPA'), axs[1, 1].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```
