### Grouping Data in a Dataframe

```python
# Group-by is a very handy Pandas function for finding statistics for different categories in a column
# for example, .count() will give the number of each value in a column that belongs to each group in the column stated.
# You can also get the mean, sum, etc of numerical columns

df.groupby(["class"]).count()
```

```python
df.groupby(["class", "odor"]).count()
```

```python
# Filter on a column

df_edible_mushys = df[df['class'] == "e"]
num_edible = len(df_edible_mushys)

display(df_edible_mushys)
num_edible
```

```python
# Filter on multiple conditions
# Use & for the AND condition, '|' for the OR condition 

df[  (df['class'] == "e") & (df['odor'] == "n") | (df['habitat'] == 'g') ]
```

```python
# Make a subset dataframe (note the double square brackets, case sensitive)

df_subset = df[ ['class', 'cap-shape', 'cap-surface']  ]
df_subset
```

```python
df
```

### Concatenate or Join Dataframes together

```python
# Concat vertically (column names are matched) axis = 0 is by rows
pd.concat([df, df_subset], axis = 1)
```

```python
# Use merge to join dataframes with a common column

# First dataframe
data1 = {'Country': ['USA', 'Canada', 'Brazil', 'India', 'China','Australia'],
         'Region': ['Americas', 'Americas', 'Americas', 'Asia', 'Asia', "Oceania"]}
df1 = pd.DataFrame(data1)
display(df1)

# Second dataFrame 
data2 = {'Country': ['USA', 'Canada', 'Brazil', 'India', 'China'],
         'Population': [331002651, 37742154, 212559417, 1380004385, 1444216107],
         'Area (sq km)': [9833517, 9976140, 8515767, 3287263, 9596961]}
df2 = pd.DataFrame(data2)
display(df2)
```

```python
# Merge the DataFrames based on the 'Country' column
merged_df =  pd.merge(df1, df2, on ='Country')

# Display the merged DataFrame
print("Merged DataFrame:")
merged_df
```
