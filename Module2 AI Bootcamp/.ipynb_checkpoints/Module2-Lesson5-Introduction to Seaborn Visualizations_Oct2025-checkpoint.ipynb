### Visualizing the distribution of a dataset

When starting with a new dataset, the first thing we want to do is get a sense for how the variables are distributed. Here we are usually takling about continuous numerical data.<br>

**distribution plots includes**: <code>displot(), jointplot(), pairplot(), rugplot(), kdeplot()</code> etc. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading data set 
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.describe()
```

```python
tips.head()
```

```python
# Visualize the distribution of the amounts for the total_bill column (histogram)
sns.displot(x = tips['total_bill'],)
plt.show()
```

```python
sns.displot(tips['size'])
```

```python
# Compare the distrubution plot of 2 features
sns.jointplot( data=tips, x='total_bill', y='tip', kind = "scatter")
```

```python
#Plot hexagons coloured to the density of points
# instead of individual points
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
# Try a ruplot! This chart plots sticks on an axis for 
# every data point on a univariate distribution
sns.rugplot(tips['total_bill'])
```

```python
# You can combine the rugplot and the distribution plot
sns.displot(tips['total_bill'], kde=False)
sns.rugplot(tips['total_bill'],  color='r')
plt.show()
```

```python
# You can create the KDE plot by changing the KDE attribute to True
sns.displot(tips['total_bill'], kde=True)
plt.show()
```

```python
# Switching to Regplot gives us the regression line!
sns.regplot(x="total_bill", y="tip", data=tips);
```

### Visualizing Categorical Data

* <code>swarmplot()</code> and <code>stripplot()</code> Plot observation for each level of the categorical variable. 
* <code>boxplot()</code> and <code>violinplot()</code> Plot an abstract representation of the distribution of variables.  
* <code>barplot()</code> and <code>pointplot()</code> Apply a statistical estimate to show a measure of central tendency and confidence interval.

```python
sns.stripplot(x="day", y="total_bill",data=tips, jitter=True, hue = "sex") # Change jitter = False for the traditional straight line
```

```python
sns.swarmplot(x="day", y="total_bill",data=tips, hue="sex")
```

```python
#Maybe you want to look at it a little differently, swap the axis!
sns.swarmplot(x="total_bill", y="day", hue="time", data=tips,);
```

```python
#Look at distributions within categories with the following plots:
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
#Can we get more information by narrowing down the categories?
sns.boxplot(x="day", y="total_bill", data=tips, hue = 'smoker')
```

```python
# Can also represent these with violin plots
# They give more information about the distribution of the data, but can be harder to interpret. Use with caution.
sns.violinplot(x="day", y="total_bill", data=tips, hue = 'smoker')
```

```python
# If we want to see the central tendancy of the categories, can try these plots:
sns.barplot(x="day", y="total_bill", hue="time",data=tips, estimator="sum") #adding "tips" displays black lines showing average of the data
```

```python
# Count plots tell us exactly how many items are in each category
# i.e. 9 Females paid on fridays, compared to 10 males.
# This is consistent for weekdays, but males pay more often on weekends
sns.countplot(x="day", hue="sex",data=tips)
```

```python
# Can automatically make subplots for a feature using catplots
sns.catplot(x="day", y="total_bill", hue="sex", col="day", data=tips, kind="bar",height=4, aspect=.75)
```

### Creating Grids of Visualizations

```python
#Remember the pairplot?
sns.pairplot(data = tips, hue="day", palette="rainbow")
```

```python
#Let's customize by colouring by sex and adding a legend using a Facetgrid. (A customizable pairplot canvas)
import matplotlib.pyplot as plt
g = sns.FacetGrid(tips, col="time",  row="smoker", hue='sex')
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

```python
# We can also make a pairgrid! 
g = sns.PairGrid(tips)
g.map(plt.scatter)
```

```python
#This allows us to customize the pairplot with specific plot types!
g = sns.PairGrid(tips)
g.map_diag(sns.histplot) # When the feature plots with itself)
g.map_upper(plt.scatter) # The grid spaces 'right' or above of the diagonal
g.map_lower(sns.kdeplot) # The grid spaces 'left' or under the diagonal
```

### Seaborn has some beautiful built in Matrix Visualizations

```python
# Plot the correlation of each pair of features and shade by the correlation value
sns.heatmap(tips.corr(numeric_only = True), annot = True) #Remember to only plot numerical columns
```

```python
# Pandas will let us create a Pivot Table
# Change the aggfunc to deterime the operation on the values falling into the grid space. 
# Mean will calculate the average for all Thursday lunches, friday lunches, Thurs Dinners, and so on.
# sum will add the values together
pvf = tips.pivot_table(values='tip', index='sex',columns='day', aggfunc='sum',observed=False)
pvf
```

```python
# You can then put the pivot table in a Seaborn heat map!
sns.heatmap(pvf)
plt.show()
```

### Seaborn also has  <code> clustermap() </code> capabilities!

It uses hierarchal clustering to produce a clustered version of a heatmap. The columns are reordered to place similar values closer together

```python
#I adjusted the pivot table to avoid the null values, which the model cannot handle
pvf = tips.pivot_table(values='tip', index='sex',columns='day', aggfunc='sum',observed=False)
g = sns.clustermap(pvf,figsize=(8,5))
```

```python
# Saturday/Sunday tips are similar and Thurs/Fri tips are similar.
# This is confirmed by the pivot table
pvf
```

### Finally, we can use Seaborn to create plots displaying Linear Regression Models

```python
# Add a line of best fit (Linear Model) to a plot with these two variables, total_bill and tip
sns.lmplot(x = 'total_bill',y = 'tip',data = tips)
```

```python
# Can display the Linear Model (lm) for multiple categories at once:
sns.lmplot(x = 'total_bill',y = 'tip',data = tips, hue = 'day')
```

```python
# And don't forget you can split out the categories by addings columns
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', col='day',)
```
