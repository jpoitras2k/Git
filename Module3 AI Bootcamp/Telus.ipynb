```python
### Load Data and pre-processing
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
scaler = StandardScaler()
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 10)


# Drop null values
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna(subset=['TotalCharges'], inplace=True)
df.drop('customerID', axis=1, inplace=True)  # ID column does not contain usefull info. axis=1 means Column wise, not row.
df.shape

### Encoding required on Churn?  Yes, from Yes and No to 0 and 1
label_encoder = LabelEncoder()
df['Churn'] = label_encoder.fit_transform(df['Churn'])

# df.head(8)

# Seperate the data for X and y
X = df.drop('Churn', axis=1,)
y = df['Churn']
# Scaling on X
X_scaled_array = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled_array, columns=X.columns)
print (X_scaled)

### View the final dataset:
#import matplotlib.pyplot as plt
#import seaborn as sn

#sn.pairplot(data = df, hue = "Churn")




```
