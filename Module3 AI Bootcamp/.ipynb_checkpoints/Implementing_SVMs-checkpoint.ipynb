Step 1
We will use SVM to train our classifier, and since we are going to perform a classification task, we will use the support vector classifier class.
We will use the iris data set, so we will import the datasets from Sklearn.
Also, we need to evaluate our model, so we will evaluate using the confusion matrix introduced in previous lessons.

```python
from sklearn.svm import SVC
from sklearn import datasets
from sklearn.metrics import confusion_matrix
```

Step 2
Now, load the data and split it into training and testing data using the sklearn.modelselection.train_test_split() the function.

```python
iris_data = datasets.load_iris()
x = iris_data.data
y = iris_data.target
x_train, x_test, y_train, y_test = 
sklearn.model_selection.train_test_split(x, y,test_size=0.2)
```

Step 3
This is the most important step.
Our data for this example is simple and linearly separable, so we will use the linear kernel.

```python
classifier = SVC(kernel='linear')
classifier.fit(x_train, y_train)
```
